```python
from datasets import load_dataset

data=load_dataset("muhammadsaadgondal/finetuningdata")
```

```python
# Extract relevant data (questions and answers) from your dataset
def get_data(dataset_dict):
    train_data = dataset_dict['train']

    # Prepare the data in the format Lamini expects (list of dictionaries)
    formatted_data = [{"input": row['question'], "output": row['answer']} for row in train_data]

    return formatted_data

# Convert your dataset to the format Lamini expects
dataset = get_data(data)

# Fine-tune the model using Lamini's API
llm.tune(data_or_dataset_id=dataset)
```

```python
llm.train(data_or_dataset_id='16f4285d5a7de5914635bb89a389ead258f15b590a3625000ca46d88391214a2')
```

```python
# Sample test question (you can pick this from your test dataset)
test_question = "What is the capital of France?"

# Use the fine-tuned model to generate an answer
response = llm.generate(test_question)

# Print the model's response
print("Question:", test_question)
print("Generated Answer:", response)
```
